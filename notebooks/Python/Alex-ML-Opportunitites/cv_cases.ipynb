```python
dfcase['DateLink'] = dfcase['ClosedDate'].values.astype('datetime64[M]')
dfcase.head()
```

```python
dfcase.groupby(['AccountId','DateLink']).size().sort_values(ascending=False)[:10]
```

```python
Casesfrom12mo = dfcase.groupby(['AccountId', 'DateLink'])['CaseNumber'].count()
Casesfrom12mo[:2]
```

```python
Casesfrom12mo = Casesfrom12mo.unstack()
Casesfrom12mo.head(3)
```

```python
Casesfrom12mo.loc[['0014000001bwjjfAAA'], :]
```

```python
#Transpose results
Casesfrom12mo = Casesfrom12mo.T
```

```python
Casesfrom12mo = Casesfrom12mo.to_period(freq='M')
```

```python
Casesfrom12mo = Casesfrom12mo.rolling(12, min_periods=1).sum()
```

```python
Casesfrom12mo['0014000001bwjjfAAA'][-10:]
```

```python
Casesfrom12mo = Casesfrom12mo.unstack(level=1)
Casesfrom12mo = pd.DataFrame(Casesfrom12mo)
```

```python
Casesfrom12mo.index.names = ['accountid', 'DateLink']
```
