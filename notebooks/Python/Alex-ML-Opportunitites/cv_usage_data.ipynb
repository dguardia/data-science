```python
rolling6MoGrAnnualized = annualizedAvgGr_BackupSize.rolling(6, min_periods=1).median()*12
BUAnnualizedGro = rolling6MoGrAnnualized.unstack()
BUAnnualizedGro = pd.DataFrame(BUAnnualizedGro)
BUAnnualizedGro.rename(columns={0:'buGroEst'}, inplace=True)
BUAnnualizedGro.head()
```

# diskspace

```python
annualizedAvgGr_Disksp = dfuse.groupby(['AccountId_sf_mrg', 'MonthStart'])['UsedDiskSpaceMB'].sum()
annualizedAvgGr_Disksp = annualizedAvgGr_Disksp.unstack()
annualizedAvgGr_Disksp = annualizedAvgGr_Disksp.T
annualizedAvgGr_Disksp = annualizedAvgGr_Disksp.to_period(freq='M')
annualizedAvgGr_Disksp = annualizedAvgGr_Disksp.pct_change()
rolling6MoGrAnnualized = annualizedAvgGr_Disksp.rolling(6, min_periods=1).median()*12
diskAnnualizedGro = annualizedAvgGr_Disksp.unstack()
diskAnnualizedGro = pd.DataFrame(diskAnnualizedGro)
diskAnnualizedGro.rename(columns={0:'diskGroEst'}, inplace=True)
diskAnnualizedGro.head()
```
